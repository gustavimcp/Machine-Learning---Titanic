## EDA

```python
eda = combined.copy()
eda['Title'] = eda['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
eda.head()
```

## 

```python
combined.drop(['Cabin', 'Ticket', 'PassengerId', 'Embarked', 'Name'], axis=1, inplace= True)
#dropping this columns because most of it is null and because we won't use it
```

```python
combined[combined['Fare'].isnull()]
```

```python
combined['Fare'] = combined['Fare'].fillna(combined.loc[(combined['Pclass'] == 3) & (combined['Sex'] == 'male') 
                                                        & (combined['Age'] < 65) & (combined['Age'] > 55)].dropna()['Fare'].mean())
```

```python
#combined[combined['Embarked'].isnull()]
```

```python
#combined['Embarked'] = combined['Embarked'].fillna(combined['Embarked'].value_counts().index[0])
#filled with letter of the place that most people embarked
```

```python
combined.reset_index(drop = True, inplace = True)
```

```python
#fill according with sex, pclass 
combined['Age'] = combined.groupby(['Sex', 'Pclass'])['Age'].apply(lambda x: x.fillna(x.median()))
```

```python
combined.isnull().sum()
```

```python
combined.head()
```

```python
combined['Familysize'] = combined['SibSp'] + combined['Parch'] + 1
combined.head()
```

```python
combined.drop(['SibSp', 'Parch'],axis= 1, inplace= True)
combined.head()
```

```python
# Changing Sex column to numbers
sex = {'male': 0, 'female': 1}
combined['Sex'].replace(sex, inplace = True)
combined.head()
```

```python
#Create 5 ranges of age
pd.cut(combined['Age'],5)
```

```python
# Discretizing the age column
combined.loc[(combined['Age'] < 16.136), 'Age'] = 1
combined.loc[(combined['Age'] > 16.136) & (combined['Age'] < 32.102), 'Age'] = 2
combined.loc[(combined['Age'] > 32.102) & (combined['Age'] < 48.068), 'Age'] = 3
combined.loc[(combined['Age'] > 48.068) & (combined['Age'] < 64.034), 'Age'] = 4
combined.loc[(combined['Age'] >= 64.034), 'Age'] = 5
combined['Age'].unique()
```

```python
max(combined['Fare'])
```

```python
pd.cut(combined['Fare'], 3)
```

```python
#Discretizing the column Fare
combined.loc[(combined['Fare'] < 88.792), 'Fare'] = 1
combined.loc[(combined['Fare'] >= 88.792) & (combined['Fare'] < 175.583),'Fare'] = 2
combined.loc[(combined['Fare'] > 175.583), 'Fare'] = 3
combined['Fare'].unique()
```

```python
combined.head()
```

```python
combined['Familysize'].value_counts()
```

```python
pd.set_option('display.max_columns', 30)
combined_2 = pd.get_dummies(columns = ['Pclass','Age', 'Fare', 'Familysize'], data= combined)
combined_2.head()
```

```python
#splitting into train and test dataframes
train_df = combined_2[:trainrow]
test_df = combined_2[trainrow:]
train_df.head()
```

```python
x = train_df.drop('Survived', axis= 1)
y = train_df['Survived']
x_pred = test_df.drop('Survived', axis= 1)
y_pred = test_df['Survived']
```

## Adjusting the Machine Learning Model

```python
#Adjusting the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

depth = [3, 5, 9, 11]
criterion = ['gini', 'entropy']
sample_splits = [2, 4, 6, 8]
sample_leaf = [1, 2, 4 ,6]
values = {'max_depth': depth, 'criterion': criterion, 'min_samples_split': sample_splits, 'min_samples_leaf': sample_leaf}

model = RandomForestClassifier(n_jobs= -1)

rscv = RandomizedSearchCV(estimator = model, param_distributions= values, n_jobs= -1)
rscv.fit(x,y)

print('Max Depth:',rscv.best_estimator_.max_depth)
print('Criterion:',rscv.best_estimator_.criterion)
print('Minimum split:',rscv.best_estimator_.min_samples_split)
print('Minimum leaf:',rscv.best_estimator_.min_samples_leaf)
print('R:', rscv.best_score_)
```

```python
#second adjustment
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

depth = [9, 11, 13, 15]
criterion = ['gini', 'entropy']
sample_splits = [6, 8, 10, 12, 18]
sample_leaf = [1, 2, 4 ,6]
values = {'max_depth': depth, 'criterion': criterion, 'min_samples_split': sample_splits, 'min_samples_leaf': sample_leaf}

model = RandomForestClassifier(n_estimators = 1000,n_jobs= -1)

rscv = RandomizedSearchCV(estimator = model, param_distributions= values, n_jobs= -1)
rscv.fit(x,y)

print('Max Depth:',rscv.best_estimator_.max_depth)
print('Criterion:',rscv.best_estimator_.criterion)
print('Minimum split:',rscv.best_estimator_.min_samples_split)
print('Minimum leaf:',rscv.best_estimator_.min_samples_leaf)
print('R:', rscv.best_score_)
```

```python
model = RandomForestClassifier(criterion= 'gini', max_depth= 13, min_samples_split= 8, min_samples_leaf= 1, n_jobs= -1)
model.fit(x, y)

predictions = model.predict(x_pred)
```
